```julia
#QUESTÃO 4.7-)
#Para conseguirmos plotar o gráfico, temos que encontrar a serie de Taylor para a função sen(x). Sabendo que o Teorema de 
#Taylor pode ser escrito como o somatório de Tn(x), começando em 0 e indo até n.
#                                              f(n)(a)(x-a)^n/n!
#Assim, queremos calcular os 5 primeiros termos desse somatório, pois vamos utilizar somente esses termos para plotar o gráfico.
#Calculando o 1 termo temos:                                            
#                                                 1*(x)^0/0! = 1
#Calculando o 2 termo (não igual a zero) temos:
#                                               -1*(x)^2/2! = - x^2/2
#Calculando o 3 termo (não igual a zero) temos:
#                                               1*(x)^4/4! =  x^4/24
#Calculando o 4 termo (não igual a zero) temos:
#                                               -1*(x)^6/6! = - x^6/720
#Calculando o 5 termo (não igual a zero) temos:
#                                               1*(x)^8/8! =  x^8/40320

#Com os termos calculados, vamos montar as funções:

#Função cos(x)
f(x) = cos.(x)

#Função com 1 termo na serie de Taylor
g(x) = 1

#Função com 2 termos na serie de Taylor
h(x) = 1 - (x^2/2)

#Função com 3 termos na serie de Taylor
i(x) = 1 - (x^2/2) + (x^4/24)

#Função com 4 termos na serie de Taylor
j(x) = 1 - (x^2/2) + (x^4/24) - (x^6/720)

#Função com 5 termos na serie de Taylor
k(x) = 1 - (x^2/2) + (x^4/24) - (x^6/720) + (x^8/40320) 

#Plotando o gráfico
#Plotando a função cos(x)
plot(f,-6,6, c=:blue, lw=2, lab="cos()")
#Plotando a função com 1 termos
plot!(g,-6,6, c=:orange, lw=2, lab="Taylor Series - 1 terms")
#Plotando a função com 2 termos
plot!(h,-6,6, c=:green, lw=2, lab="Taylor Series - 2 terms")
#Plotando a função com 3 termos
plot!(i,-6,6, c=:red, lw=2, lab="Taylor Series - 3 terms")
#Plotando a função com 4 termos
plot!(j,-6,6, c=:purple, lw=2, lab="Taylor Series - 4 terms")
#Plotando a função com 5 termos
plot!(k,-6,6, c=:brown, lw=2, lab="Taylor Series - 5 terms")

#Ajustando o gráfico no eixo y
ylims!(-6,4)
```
