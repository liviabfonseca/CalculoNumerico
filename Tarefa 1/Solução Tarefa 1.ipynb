```julia
#QUESTÃO 2.1-)
#Queremos descobrir a raíz cubica de 43. Tendo como base a função f(x) = x^3 - 43 = 0, para implementar o método de
#Newton vamos utilizar o conceito de tangência para descobrirmos x quando a função é igual a zero. Então, iremos calcular
#a derivada da função (a taxa de variação do eixo y dividida pela taxa de variação do eixo x), assim, vamos calcular:
#                                              x2 = x1 - (f(x1)/f'(x1)).

#raiz_cubica_de_43:Função que calcula a raíz cúbica de 43 utilizando o método de Newton recebendo o número de iterações e
#chute desejados
function raiz_cubica_de_43(numero_de_iterações,chute)
    for i=1:numero_de_iterações #Looping que atualiza o valor aproximado da raiz
        chute = chute-((chute^3)-43)/(3*(chute^2)) #Calculando a tangente da função com relação ao número de interações e chute 
    end
    return chute #Retorna a aproximação da raiz cúbica de 43
end
```

```julia
#QUESTÃO 4.4-)
#Queremos aproximar a função f(1.5) = ln(1.5) utilizando o polinômio de Taylor de terceira ordem, com a = 0. Primeiramente,
#vamos determinar o polinômio de Taylor de ordem 3:

#                    f(x)-(f(a)+f'(a)(x-a)+(f''(a)(x-a)^2)/2+(f'''(a)(x-a)^3)/6) ≤ M(x-a)^4/24

#Com o polinômio em mãos, vamos encontrar as derivadas pedidas:
#                                              f(a) = ln(a)
#                                              f'(a) = 1/a
#                                             f''(a) = -1/a^2
#                                             f'''(a) = 2/a^3
#Substituindo a = 1 nas derivadas, encontramos:
#                                            f(1) = ln(1) = 0
#                                           f'(1) = 1/1 = 1
#                                           f''(1) = -1/1^2 = -1
#                                          f'''(1) = 2/1^3 = 2

#Substituindo os valores no polinômio de Taylor, onde x = 1.5 e a = 0, temos:

#                     ln(1.5)-(0 + 1 (1.5 - 0)+(-1(1.5 - 0)^2)/2+(2(1.5-0)^3)/6) ≤ M(1.5-0)^4/24
#                                  ln(1.5)-0.416666... ≤ M(0.0026041666666666665)

#Agora temos que encontrar o valor de M. Sabendo que M é definido como o modulo do valor máximo encontrado da derivada de n+1
#no intervalo de (a,x). Nesse caso, como estamos trabalhando com Taylor de ordem 3, vamos calcular a 4 derivada da função (3+1).
#Sabemos também que estamos trabalhando em um intervalo de 1 a 1.5, logo teremos que descobrir o maior valor encontrado quando
#substituirmos os valores do intervalo na 4 derivada da função.

#                                           f''''(n) = -6/(n^4)

#Como a função está sendo dividida por n, sabemos que n = 1 será o maior valor que a função irá assumir no intervalo dado. Dessa
#forma, teremos:
#                                          M = |-6/(1^4)| = 6
#Substituindo o M na fórmula:                            
#                               ln(1.5)-0.416666... ≤ 6(0.0026041666666666665)
#                                     ln(1.5)-0.416666... ≤ 0.015625

#Encontramos que a aproximação de ln(1.5) é 0.416666... e possui 1 casa decimal correta depois da virgula. Se analisarmos
#a inequação encontrada temos que a parte esquerda dela é o erro da aproximação (ln(1.5)-0.416666...) e a parte direita é o 
#teto máximo desse erro. Podemos observar que o teto possui 1 casa decimais zerada depois da virgula, o que indica que 
#essa casa é igual na aproximação e no valor "real" do ln(1.5).


###CALCULANDO O POLINÔMIO PARA O ERRO DE TRUNCAMENTO MENOR QUE 10^-8

#Agora queremos encontrar quantos termos possui o  polinômio para o erro de truncamento ser menor que 10^−8. Nesse caso, sabemos
#que para o erro atender essa especificação teremos a desigualdade:

#                                         M*((x−a)^n+1/(n+1)!) < 10^-8  

#Substituindo os valores de x = 1.5 e a = 1:
#                                        M*((0.5)^n+1/(n+1)!) < 10^-8  

#Porém, temos que encontrar a fórmula para um M genérico. Assim, sabendo que M é definido como o módulo do valor máximo
#encontrado da derivada de n+1 no intervalo de (1,1.5), vamos analisar as derivadas da função:

#                                               f(k) = ln(k)
#                                              f'(k) = 1/k
#                                             f''(k) = -1/k^2
#                                            f'''(k) = 2/k^3
#                                           f''''(k) = -6/(k^4)
#                                            f5(k) = 24/(k^5)
#                                                   ...

#Podemos observar que as derivadas possuem um padrão:

#                                        (((-1)^(n2+1))*(n2-1)!)/k^n2

#Com isso, vamos reescrever nossa fórmula geral para M, retirando a parcela (-1)^(n2+1), visto que não estamos nos preocupando 
#com a possibilidade de possuirmos uma valor negativo, pois o 'M' é o módulo do valor encontrado:

#                                                 (n2-1)!/k^n2
#Podemos simplificar essa expressão, pois sabemos que o nosso a = 1, logo a parcela que divide a expressão (k^n2) sempre terá
#valor igual a 1.Logo, ficamos somente com:
#                                                    (n2-1)!
#Sabendo que a derivada que estamos calculando no M é a derivada do próximo termo, temos que n1 = n + 1, sendo n o número "atual"
#da derivada:
#                                                   (n2-1)!
#                                                    (n)!

#Calculando a desigualdade, temos:   

#                                       M*((0.5)^n+1/(n+1)!) < 10^-8   
#                                      n! * ((0.5)^n+1/(n+1)!) < 10^-8 

#Sabendo que  (n+1)! = n!*(n+1), podemos simplificar a inequação:                                             
#                                   n! *((0.5)^n+1/n!*(n+1)) < 10^-8
#                                      ((0.5)^n+1/(n+1)) < 10^-8                             

#Substituindo os valores de x = 1.5 e a = 1:
#                                        ((0.5)^n+1/(n+1)) < 10^-8

#Com a fórmula já pronta, vamos chutar alguns valores para n até que nossa condição seja atendida:
# n = 10:
# ((0.5)^10+1/(10+1)) < 10^-8 #Condição não atendida

# n = 19:
# ((0.5)^19+1/(19+1)) < 10^-8 #Condição atendida

#Dessa forma, temos que o  polinômio para o erro de truncamento ser menor que 10^−8 possui 19 termos.
```

```julia
#QUESTÃO 4.5-)
#Vamos criar um programa que calcula o ln(x) dado um valor para o erro máximo 
#Função calcula_ln: ela recebe o valor de 'x' que é o ln que ele deseja calculae e o valor de 'erroMax' que é o valor do erro 
#máximo aceitado por ele.
function calcula_ln(x,erroMax)
    n = 1 #contador da função while
    s = 1.0 #varivavel que irá armazenar o valor do ln de x
    while ((((x)^(n+1))/(n+1)) > (erroMax)) #Lopping for para enquanto o valor do erro for menor maior que o aceitável
        s+=x^n/factorial(n) #somando mais uma parcela ao valor aproximado
        n = n + 1 #somando 1 ao contador
    end
    return  s #retorna o valor do ln
end
```
